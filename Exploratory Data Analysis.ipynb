# Exploratory Data Analysis (EDA) - NYC Department of Education Quality Review and Surveys

**Import Packages**

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
sns.set_style('white')
import matplotlib.pyplot as plt
%matplotlib inline
```

**Read Data**

```python
nyc_schools = pd.read_csv('Data/nyc_schools.csv')
nyc_schools.head()
```

```python
nyc_schools.describe()
```

**Distributions of Borough and District**

```python
fig = plt.figure(figsize=(15, 5))
plt.suptitle(f'School Neighborhoods \n', size=16)
for row, area in enumerate(['Borough', 'District']):
    plt.subplot(1, 2, row+1)
    sns.countplot(nyc_schools[area],
                  palette='hls')
    sns.despine(left=False, bottom=False)
    if row == 1:
        plt.xticks(rotation=90)
plt.savefig('Images/neighborhoods.png',
            bbox_inches='tight')
```

**Distributions of Quality Review Ratings and Survey Scores**

```python
ratings = [nyc_schools.columns[3]] + \
    list(nyc_schools.columns[5:15]) + list(nyc_schools.columns[-5:-2])
```

```python
fig = plt.figure(figsize=(15, 15))
plt.suptitle(f'Quality Review Ratings', size=16)
for row, rating in enumerate(ratings[:-5]):
    plt.subplot(3, 3, row+1)
    sns.countplot(nyc_schools[rating],
                 palette='husl')
    sns.despine(left=False, bottom=False)
plt.subplots_adjust(wspace=0.3, hspace=0.3);
```

```python
fig = plt.figure(figsize=(15, 10))
plt.suptitle(f'Survey Scores', size=16)
for row, rating in enumerate(ratings[-5:]):
    plt.subplot(2, 3, row+1)
    sns.distplot(nyc_schools[rating])
    sns.despine(left=False, bottom=False)
plt.subplots_adjust(hspace=0.3);
```

**Distributions of School Demographics**

```python
school_dem = ['Enrollment', 'Percent English Language Learners',
              'Percent Students with Disabilities',
              'Percent Self-Contained', 'Economic Need Index',
              'Percent in Temp Housing', 'Percent HRA Eligible',
              'Percent Asian', 'Percent Black',
              'Percent Hispanic', 'Percent White',
              'Principal experience at this school',
              'Percent of teachers with 3+ years of experience',
              'Student Attendance Rate',
              'Percent of Students Chronically Absent',
              'Teacher Attendance Rate']
```

```python
fig = plt.figure(figsize=(20, 25))
plt.suptitle(f'School Demographics', size=16)
for row, demog in enumerate(school_dem):
    plt.subplot(4, 4, row+1)
    sns.distplot(nyc_schools[demog])
    sns.despine(left=False, bottom=False)
plt.subplots_adjust(hspace=0.3);
```

## (Classification)  How does a school's environment affect its overall student achievement rating?

Predict whether a school will not meet (1), approach (2), meet (3), or exceed (4) target

**Distributions of Student Achievement Ratings**

```python
vc = nyc_schools['Student Achievement Rating'].value_counts()/len(nyc_schools)
sns.barplot(x=vc.index, y=vc, palette='hls')
sns.despine(left=False, bottom=False)
plt.title('Student Achievement Distribution \n')
plt.xlabel('Student Achievement Rating')
plt.ylabel('Percent')
plt.savefig('Images/student_achiev_dist.png',
            bbox_inches='tight')
```

**Distribution of Student Achievement Rating by Borough**

```python
df_borough = nyc_schools.groupby(['Student Achievement Rating',
                                  'Borough']).size().reset_index().pivot(columns='Borough',
                                                                         index='Student Achievement Rating',
                                                                         values=0)
df_borough.plot(kind='bar', stacked=True,
                figsize=(7, 5), colormap='YlGnBu')
sns.despine(left=False, bottom=False)
plt.title('Boroughs by Student Achievement Rating \n')
plt.ylabel('Count')
plt.legend(loc='upper left')
plt.savefig('Images/boroughs_achiev_rating.png',
            bbox_inches='tight')
```

**Distributions of Quality Review Ratings by Student Achievement Ratings**

```python
fig = plt.figure(figsize=(20, 20))
plt.suptitle(f"Student's achievement and quality review ratings", size=16)
for row, rating in enumerate(ratings):
    plt.subplot(4, 4, row+1)
    sns.boxplot(x=nyc_schools['Student Achievement Rating'],
                y=nyc_schools[rating],
                order=[1, 2, 3, 4],
                palette='Paired')
    sns.despine(left=False, bottom=False)
plt.subplots_adjust(wspace=0.4, hspace=0.4);
```

**Distributions of School Demographics by Student Achievement Ratings**

```python
fig = plt.figure(figsize=(20, 20))
plt.suptitle(f'School Demographics', size=16)
for row, demog in enumerate(school_dem):
    plt.subplot(4, 4, row+1)
    sns.boxplot(x=nyc_schools['Student Achievement Rating'],
                y=nyc_schools[demog],
                order=[1, 2, 3, 4],
                palette='Set2')
    sns.despine(left=False, bottom=False)
plt.subplots_adjust(wspace=0.4, hspace=0.4);
```

```python
plt.figure(figsize=(20, 20))
sns.set(font_scale=1)
sns.heatmap(nyc_schools.corr(), cmap='RdYlGn', annot=True)
plt.xticks(fontsize=13, rotation=50, ha='right')
plt.yticks(fontsize=13)
plt.title("Correlation B/W Different Features", fontsize=18);
```

## (Regression) How does a school's environment affect its students' ELA Assessment scores?

Predict percentage of students who will receive 3+ on the ELA assessment

```python
nyc_schools_ela = pd.read_csv('Data/nyc_schools_ela.csv')
nyc_schools_ela.head()
```

```python
# plot distribution of ELA scores
sns.distplot(nyc_schools_ela['Level 3+ %'])
sns.despine(left=False, bottom=False)
plt.xlim(0, 100)
plt.title('Students with 3+ ELA Scores \n')
plt.savefig('Images/3+_ela.png',
            bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(12, 10))
plt.suptitle(f'ELA Score Distribution', size=16)
for row, score in enumerate(nyc_schools_ela.columns[-6:-2]):
    plt.subplot(2, 2, row+1)
    sns.distplot(nyc_schools_ela[score])
    sns.despine(left=False, bottom=False)
plt.subplots_adjust(wspace=0.3, hspace=0.3);
```
